```python
import seaborn as sns

plt.figure(figsize=(8, 5)) 
ax = sns.countplot(data=df, x='Attrition', palette='Set2')  
ax.set_title('Employee Attrition Count', fontsize=16)
ax.set_xlabel('Attrition Status', fontsize=12)
ax.set_ylabel('Count', fontsize=12)

sns.despine() 
plt.tight_layout() 
plt.show()
```

```python
plt.figure(figsize=(9, 6))
sns.set_style('whitegrid') 

ax = sns.boxplot( data=df, x='Attrition', y='MonthlyIncome', palette='pastel',  
     showmeans=True, meanprops={  "marker": "D",  "markerfacecolor": "red", "markeredgecolor": "black", "markersize": 7 })
ax.set_title('Monthly Income Distribution by Attrition Status', fontsize=16)
ax.set_xlabel('Attrition', fontsize=12)
ax.set_ylabel('Monthly Income', fontsize=12)

sns.despine() 
plt.tight_layout()
plt.show()
```

```python
sns.set(style='whitegrid')  
plt.figure(figsize=(9, 6)) 
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

ax = sns.barplot(data=df, x='OverTime', y='Attrition_Flag', palette='Blues_d', ci=None)
ax.set_title('Attrition Rate by Overtime Status', fontsize=16, fontweight='bold')
ax.set_xlabel('Overtime', fontsize=12)
ax.set_ylabel('Proportion of Employees Who Left', fontsize=12)

sns.despine()  
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14,10))
df_encoded = pd.get_dummies(df,drop_first=True)
sns.heatmap(df_encoded.corr(),cmap='coolwarm',annot = False)
plt.title("correlation matrix")
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))

ax = sns.histplot(data=df,x='Age',hue='Attrition',kde=True,bins=30,palette='Set2',edgecolor='white',alpha=0.6)
ax.set_title('Age Distribution by Attrition Status', fontsize=16, fontweight='bold')
ax.set_xlabel('Age', fontsize=12)
ax.set_ylabel('Employee Count', fontsize=12)
median_age = df['Age'].median()
plt.axvline(median_age, color='gray', linestyle='--', linewidth=1)
plt.text(median_age + 1, ax.get_ylim()[1]*0.9, f'Median Age: {median_age:.1f}',
         color='gray', fontsize=10)

sns.despine()
plt.tight_layout()
plt.show()
```
